```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def white_balance_grayworld(image):
    avg_color_per_row = np.average(image, axis=0)
    avg_color = np.average(avg_color_per_row, axis=0)
    norm_avg_color = np.sum(avg_color) / 3.0
    white_balanced = image * (norm_avg_color / avg_color)
    return np.clip(white_balanced, 0, 255).astype(np.uint8)
```

```python
def sharpen_image(image):
    blurred = cv2.GaussianBlur(image, (0, 0), 3)
    sharpened = cv2.addWeighted(image, 1.5, blurred, -0.5, 0)
    return sharpened
```

```python
image_path = '/kaggle/input/ikan-bawah-laut/underwater_fish.jpg'
image = cv2.imread(image_path)
```

```python
white_balanced_image = white_balance_grayworld(image)
sharpened_image = sharpen_image(white_balanced_image)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(cv2.cvtColor(white_balanced_image, cv2.COLOR_BGR2RGB))
plt.title('White Balanced Image')
plt.axis('off')

plt.subplot(133)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')
plt.axis('off')

plt.show()
```
